```python
import pandas as pd
import numpy as np
df = pd.read_csv('games.csv')
df.head()
```

```python
#Opeartions on Pandas
```

```python
#Check basic info See column names, non-null values, and data types:
df.info()
```

```python
#Get dataset statistics Useful for numeric columns (like ratings, turns, etc.):
df.describe()
```

```python
#Check how many games each player played as white (if there’s a column like white_id)
df['white_id'].value_counts().head()
```

```python
#Filter: Get games where white won (if there's a winner column)

white_wins = df[df['winner'] == 'white']
white_wins.head()
```

```python
#Count how many games each opening was used:

df['opening_name'].value_counts().head()
```

```python
#Plot average number of turns by winner:

df.groupby('winner')['turns'].mean()
```

```python
# Operations on Numpy
```

```python
#1. Calculate average number of turns using NumPy

avg_turns = np.mean(df['turns'])
print("Average number of turns:", avg_turns)
```

```python
#2. Find the standard deviation of turns

std_turns = np.std(df['turns'])
print("Standard deviation of turns:", std_turns)
```

```python
#3. Add a new column: "is_long_game" using NumPy where turns > 60

df['is_long_game'] = np.where(df['turns'] > 60, True, False)
df[['turns', 'is_long_game']].head()
```

```python
#Find max and min ratings (if rating columns exist like white_rating)

max_white_rating = np.max(df['white_rating'])
min_white_rating = np.min(df['white_rating'])

print("Max white rating:", max_white_rating)
print("Min white rating:", min_white_rating)
```

```python
#5. Count how many games lasted exactly median number of turns

median_turns = np.median(df['turns'])
count_median = np.sum(df['turns'] == median_turns)

print(f"Games with median turns ({median_turns}):", count_median)
```
