```python
import numpy as np

X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

Y = np.array([[0],
              [0],
              [0],
              [1]])

# ----------------------------
# Initialize weights
# ----------------------------
input_size = X.shape[1]
output_size = Y.shape[1]

np.random.seed(42)
W = np.zeros((input_size, output_size))  # initialize weights to 0

# Learning rate
eta = 0.1

# ----------------------------
# Hebbian Learning
# ----------------------------
for epoch in range(1, 6):  # train for 5 iterations
    print(f"Epoch {epoch}")
    for x, y in zip(X, Y):
        # Compute weight change
        delta_W = eta * np.outer(x, y)
        # Update weights
        W += delta_W
        print(f"Input: {x}, Target: {y}, Weight change:\n{delta_W}\nUpdated Weights:\n{W}\n")
```
